## What is Research?
* **Research** is a systematic process of collecting, analyzing, and interpreting information (data) in order to increase our understanding of a phenomenon about which we are interested or concerned
    * collection of data and **interpretting** it
* Research is **cyclical**
    1. begin with an unanswered problem
    2. clearly and specifically articulates the goal of the research
        * define the **research problem**
    3. divide the principle problem into more managable subproblems
    4. identify assumptions and hypotheses
    5. specify a plan for addressing the problem and subproblems
    6. collect, organize, and analyze data related to the problem and subproblems
    7. interpret the meaning of the data as related to the research
        * the most important parto fthe research. the researcher extracts *meaning* from the data
        * interpretation is a subjective process that depends on the researchers assumptions
* Research is **iterative**
    * a researcher may need to move back and forth between research steps
* a **Hypotheses** is a logical supposition, and eduucated guess, providing a tentative explanation for the phenomenon under investigation
    * **a priori hypotheses** are proposed prior to data collection
    * good to begin research with an open mind
* a **research plan** is a detailed step by step itinerary of the route a researcher intends to take to answer the research problem
* The two general forms of Research
    * can take one or both
    * **Quantitative Research** involves looking at amounts, or quantities, of one or more vairables of interest
    * **Qualitative Research** involves looking at characteristics, or qualities, that cannot easily be reduced to numerical values
    * both quantitative and qualitative research require distinctly different research methods and data analysis strategies
* other forms of research
    * **Mixed-Methods Research** uses both quantitative and qualitative
    * **Action Research** putting the research hypothesis in question in action (mainly social research? chapter 10)

## Philosophical Assumptions Underlying Research Methodologies
* two general assumptions that underly many research studies
    * the phenomenon under investigation is somewhat lawful and predictable, and it is *not* comprised of completely random events
    * cause-and-effect relationships can account for certain patterns observed in the phenomenon
* philosophies that govern research
    * **positivism** believes that with the appropriate measurement tools, scientists can objectively uncover absolute, undeniable *truths* about cause-and-effect relationships
    * **postpositivism** believes that true objectivity is hard to perform. researchers inevitably bring **biases** into their research
        * progress towards understanding tends to be gradual and probabilistic
        * hypotheses are either **supported** or **not supported** by data
    * **constructivism** believes that bias is unavoidable in research and try to be as up-front about their possible biases as possible
        * focuses on peoples *perceptions* and *interpretations*
    * **phenomenology** focuses on how human beings experience themselves and the world
        * questions usually take the form of "what is it like to experience such-and-such?"

## The Human Mind

### Critical Thinking
* involves evaluating the accuracy, credibility, and worth fo information and lines of reasoning
* forms
    * **verbal reasoning** understanding and evaluating persuasive techniques found in language
    * argument analysis - discriminating between reasons that do and do not support a particular conclusion
    * **probabilistic reasoning** -determining the likelihood and uncertainties associated with various events
    * **decision making** - identifying and evaluating several alternatives to the proposed cause
    * **hypothesis testing** - judging the value of data nd research results in terms of the methods used to obtain them and their potential relevance to certain conclusions

### Deductive Logic
* begin with one or more **premises** and proceeds logically from those premises to a conclusion
* premises may be flawed

### Inductive Reasoning
* begins with one or more observations and draw conclusions about entire classes of objects or events
* cannot rule out exceptions or anomalies, but the conclusion is supported by the observed data

### Scientific Method
* systematic way of collecting and analysing data
* steps
    1. identify a problem
    2. create a hypothesis that potentially resolves the problem (deductive?)
    3. gather data relevant ot the hypothesis (inductive?)
    4. analyze and interpret the data to see whether they support the hypothesis
* many researchers do not rigidly follow the scientific method. should be used as a guide
* uses both deductive and inductive reasoning

### Theory Building
* a **theory** is an organized body of concepts and principles intended to explain a particular phenomenon
* as new data emerges a theory may have to be revised, or even rejected (in which case an alternative theory is proposed)

# The Problem: The Heart of the Research Process

## Finding Research PRojects
* **Basic Research** are research projects that can advance theoretical conceptualizations about principle topics
* **Applied Research** are research projects that have immediate relevance to current procedures, practices, and policies (Handwriting Identification is probably this)
* four situations to avoid when considering a research process
    1. should not be only for self enlightnemnet. should have a benefit
    2. should not solely compare two sets of data
    3. simply calculating a correlation coefficient is not acceptable research
        * need to make sensse of the data
        * correlation is not causation
    4. a problem that results in a yes or no answer is not sufficient for a research study
* Choosing an Appropriate Problem or Question
    1. find gaps in knowledge in your field/area of interest
    2. read the existing research literature about a topic
        * keep a running record of journal articles
    3. Seek the advice of experts
    4. attend professional conferences
    5. choosing a topic that you are interested in
    6. choose a topic that other individuals will find interesting and worthy of attention
    7. be realistic about what you can accomplish
* defining a **Purpose Statement** - a statement that is carefully phrased adn represents one or more goals of the total research effort
    1. describe your problem clearly and completely
    2. think through the feasibility of the project that your purpose statement implies
    3. say precisely what you mean
    4. state the problem is a way that reflects an open mind about its solution

    

## Subproblems Vs. Pseudo-Subproblems
* subproblems are integral parts of the main problem
* **pseudo subproblems** are things that look like subproblems, but are actually *procedural issues*
    * examples
        * qualitative, quantitative, or mixed-methods for research?
        * how to choose sample from larger population
        * specific methods for data collection
        * procedures for data analysis

## Characteristics of Subproblems
1. each subproblem should be a research problem in its own right
    * a logical subdivision that can be a ditinct project in the larger research endeavor
2. each subproblem must be clearly tied to the interpretation of data
3. the subproblems must add up to the totality of the subproblem
    * should address all significant aspects of the main problem
4. subproblems should be small in number
    * two to six

## Identifying Subproblems
* paper-and-pencil approach
    1. write problem down with room to write
    2. critically examine the problem to identify specific topics that require in-depth treatment
    3. ensure that the text within each box includes one or more words implying the need for data interpretation
    4. arrange the entire problem or question - which now has subproblems in boxes - to show the structure of the whole research design
* Mind Mapping
    * use software to construct a graphical network of interrelated concepts, terms, and principles

## Identifying a Relevant Theoretical or Conceptual Framework
* a **Theoretical Framework** is a pre-existing or newly proposed *set of concepts, definitions, and principles* that, in combination, may shed light on cause-and-effect relationships
    * provides a *lens* for an issue
    * may provide the basis for one or more *a priori* hypotheses
* a **Conceptual Framework** identifyes key concepts and concept connections that provide a helpful *lens* through which to look at certain processes
    * typically visual representation of the expected realtionships and connections
* secondary source: grad coach

## Stating Hypotheses
* Hypotheses are intelligent, reasonable, and testable, guesses
* often a one-to-one correspondence to subproblems and hypotheses
* *a priori* hypotheses are ultimately *tenetive propositions to assist or guide the investigation of a research problem*

### Research Hypotheses and Null Hypotheses
* a **research hypotheses** are as defined above
* a **null hypothesis** is a hypothesis opposite of the proposed research hypothesis (hypotheses)
    * a researcher may sseek to *reject* (cast doubt on) the null hypothesis
    * typically a "no difference" hypothesis
    * better suited for statistical analysis than as an *a priori* hypothesis
    * example:
        * research hypothesis: Teenagers enrolled in Program A will graduate from high school at a higher rate than teenagers enrolled in Program B
        * null hypothesis: there will be no difference in the highschool graduation rates of teenagers enrolled in Program A and Program B

## Identifying the General and Possibly Also Specific Variables that are the Focus of the Investigation
* general concepts must be identified, *at minimum*
* a **variable** is any characteristic, experience, behavior, or outcome in a research investigation that has two ormore possible values
    * concept and variable may overlap. quantitative research tends to use *variable* while qulaitative research tend to let concepts and variable emerge
* cause-and-effect relationships typically have
    * one or more **independant variables** which the researcher manipulates
    * one or more **dependent variables** which may be influenced by the indepdendant variables
    * typically: Independent ==> Dependent
    * a **mediating Variable** (AKA an **intervening variable**) may help explain why a variable has the effect it does on the indepdenent variable
        * Independent ==> Mediating ==> Dependent
    * a **moderating vairable** is a variable that, while not intervening, influences the nature and strength fo the cause-and-effect realtionship
        * distinct because an independent variable may potentially influence a mediating variable, but not a moderating variable

## Defining Terms

* avoid circular definitions
* an **operational definition** is a definition the researcher defines in terms of how it wil be identified and assessed in the study

* some assumptions are so basic that without them, the research problem would not exist

## Knowing When to Quit
* if you start seeing repetitive patterns in teh material you are finding and reading then it may be time to bring the literature review to a close
* may need to resume when you find gaps in your knowledge of a subject
* you shouldn't omit or misrepresent prior research findings

## Writing the Literature Review
* writing a clear and cohesive literature review
    1. get the proper psychological orientation
        * be clear in your thinking
        * literature review as a description for a peer about what other people have written on the subject
    2. develop a plan for the overall organizational structure of your review
        * create an outline of topics you intend to address and the points you intend to make
        * discuss the literature broadlya t first and then go into more detail
    3. continually empohasize relatedness to your research problem
        * should not be a chain of isolated summaries of other peoples research
    4. provide transitional phases or paragraphs to help readers follow your rain of thought
    5. describe litearture, not plagiaraize it
        * **plagiarism** involes
            * presenting another person's work as being your own
            * insufficientlya cknoledging and identifyign the sources from which one has drawn in writing
            * paraphrase as much as you can
    6. always give credit where it is due
    7. minimize the use of direct quotes
    8. summarize what you have said (at the end)
    9. remember that your first draft will *not* be the last draft
    10. get continual feedback

# Chapter 4 - Planning your Research Project
* a **research design** is a general strategy/plan and provides the overall structure for
    * the procedure the researcher follows
    * the data the researcher collects
    * the data analysis the researcher conducts

## Planning a General Approach
* problems arise out of broad, generic areas
    * people: problems related to human beings, phychology, learning, motivation, social interaction, crime, rehabilitation, medical treatments, language, etc
    * things: objects and non-hman entities
    * records: problems related to newspapers, journals, letters, the internet, websites, registers, speeches, legal documents, etc
    * dynamics and energy: related to human interactions, chemical reactions, radiation, etc
* a **unit of analysis** is "who", or "what", is the focus of the study (handwriting)

## Research Planning vs. Methodology
* a *methodology* is more specific than a *plan*
    * depends on the
        * discipline
        * research problem
        * the kind of data that needs to be collected and analyzed
* a research plan should detail the types of data to be collected as well as the method of collection


## The Nature and Role of Data in Research
* research is collecting and interpreting relevant data
* data is not an absolute reality or truth. can be unrepresentative of the world as a whole
* data is **transient**
    * data are rarely permanent, and may be accurate for only a short time

## Primary vs. Secondary Data
* **Primary Data** is the most valid and relevant information collected in a research project. derived directly from a research project
* **Secondary Data** is derived from primary data
    * inevitable bias and abstraction of detail

## Planning for Data Collection
1. What Data are Needed?
2. Where is the Data Located
3. How will the Data be obtained
    * may be subject to law, confidentiality, ethics
4. What limits will be placed on the nature of acceptable data
    * not all gathered data will be acceptable
    * may have to set criteria and limits on the gathered data
    * **criteria for the admissibility of data** is the set of identified restrictions
5. How will the Data be Interpreted
    * look aback at the research problem

## Linking Data and Research Methodology
* to some extent, the desired data dictates the methodology

### Comparing Quantitative and Qualitative Methodologies
* quantitative and qualitative process share some similar processes
    * identifying a research problem
    * reviewing related literature
    * collecting/analyzing data
* Quantitative and Qualitative processes are more suitable for different types of data
    * quantitative deals with numerical/ordinal data
    * qualitative deals with nonnumerical and unstructured data
* Typical characteristics of quantitative vs. qualitative

* purpose
    * Quantitative researchers tend to seek explanations and predictions that will generalize to other persons and places. with the intent to identify relationships among two or more variables
    * Qualitative researchers tend to seek better understandings of complex situations. tends to be exploratory in nature
* Process
    * Quantitative research is generally very structured. concepts, variables, hypothesees, and assessment techniques tend to be defined before the stufy begins. methods are chosen that allow obective measurement of the variables of interest
    * Qualitative research is more holistic and emergent. the specific focus, design, and data collection method developes and changes along the way
