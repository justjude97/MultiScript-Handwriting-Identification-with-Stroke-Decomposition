```python
import numpy as np
import matplotlib.pyplot as plt

import cv2 as cv
from skimage.morphology import skeletonize, remove_small_holes
import networkx as nx

#module for converting skeleton networks into a graph representation
import sknw.sknw as sknw
import qdanalysis.strokedecomposition as sd
from qdanalysis.preprocessing import preprocess

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
cerug = cv.imread(r'CERUG_EN_TEST.png', cv.IMREAD_GRAYSCALE)
```

```python
cerug_proc = preprocess(cerug)
plt.imshow(cerug_proc)
plt.show()
```

```python
cerug_skele = skeletonize(cerug_proc)
plt.imshow(cerug_skele)
```

```python
cerug_junctions = sknw.mark_node(cerug_skele) == 2
plt.imshow(cerug_junctions)
```

```python
cerug_graph = sknw.build_sknw(cerug_skele, multi=True, full=True)
```

```python
image = np.zeros_like(cerug)
sknw.draw_graph(image, cerug_graph)
```

```python
plt.imshow(image)
plt.show()
```
